```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq

seq_file = "gbbct359.seq"

sequences = list(SeqIO.parse(seq_file, "genbank"))
```

```python
# Question no.1
for sequence in sequences:
   
    print("ID:", sequence.id)

print("Total number of records:", len(list(sequences)))
```

```python
# Question no.2

print("ID  \t\t  GC Factorization")
for sequence in sequences:
    gc = gc_fraction(sequence.seq.upper())
    print(sequence.id, "\t",gc)
```

```python
# Question no.3

template_strand = []
for sequence in sequences:
    ts = sequence.seq.reverse_complement()
    template_strand.append(ts)
template_strand
```

```python
# Question no.4

mRNA_sequence = []
for sequence in sequences:
    mRNA = sequence.seq.transcribe()
    mRNA_sequence.append(mRNA)
    
mRNA_sequence
```

```python
# Question no.5

proteins = []
for mrna in mRNA_sequence:
    protein = mrna.translate()
    proteins.append(protein)


proteins
```

```python
protein_records = [
    SeqIO.SeqRecord(
        seq,
        id=f"protein_{i + 1}",
        description=f"Protein {i + 1}",
        annotations={"molecule_type": "protein"}
    )
    for i, seq in enumerate(proteins)
]

# Specify the path to your output GenBank file
output_genbank_file = "output_proteins.gb"

# Write the protein SeqRecords to a GenBank file
with open(output_genbank_file, "w") as output_handle:
    SeqIO.write(protein_records, output_handle, "genbank")

print(f"Proteins translated and saved in {output_genbank_file}.")






```
