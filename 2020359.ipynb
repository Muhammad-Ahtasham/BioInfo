```python
# Question no.1
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq

seq_file = "gbbct1004.seq"

sequences = list(SeqIO.parse(seq_file, "genbank"))
```

```python
# Question no.2
for sequence in sequences:
   
    print("ID:", sequence.id)
    print("Description:", sequence.description)

print("Total number of records:", len(list(sequences)))
```

```python
#Question no.3
for sequence in sequences:
    print("ID:", sequence.id)
    
    # Iterate through the features in the record
    for feature in sequence.features:
        feature_type = feature.type  # Get the feature type
        feature_location = feature.location  # Get the feature location
        
        print("Feature Type:", feature_type)
        print("Feature Location:", feature_location)
       
    # Print a separator line to separate records
    print("-" * 40)
```

```python
# Question no.4
# Iterate through each record in the list
for i, sequence in enumerate(sequences):
    print("Searching in Record ID:", sequence.id)
    
    # Iterate through the features in the current record
    for feature in sequence.features:
        feature_location = feature.location  # Get the feature location
        feature_seq = feature.extract(sequence.seq)  # Extract the feature sequence
        
        # Iterate through the other records to search for matching feature sequences
        for j, other_sequence in enumerate(sequences):
            if i != j:  # Skip comparing a record to itself
                other_sequence_id = other_sequence.id
                other_sequence_seq = other_sequence.seq
                
                # Check if the feature sequence exists in the other record's sequence
                if feature_seq in other_sequence_seq:
                    # Find the location of the feature in the other record's sequence
                    start = other_sequence_seq.find(feature_seq)
                    end = start + len(feature_seq)
                    
                    print(f"Match found in Record ID: {other_sequence_id}")
                    print(f"Location in {other_sequence_id}: {start + 1}-{end}")
    
    # Print a separator line to separate records
    print("-" * 40)
```
